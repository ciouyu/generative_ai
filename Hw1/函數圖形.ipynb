#黃金螺旋#
IMG_5486.jpeg


意即   
我要實作黃金比例(1+sprt(5)/2）   
對phi取log/90度獲得成長率b   
初始半徑為a 乘上exponential的b*theta次方。
剛好numpy都有以下功能

```python
import numpy as np
import matplotlib.pyplot as plt

# 黃金比例
phi = (1 + np.sqrt(5)) / 2

# 計算成長率 b
b = np.log(phi) / (np.pi / 2)

# theta 角度（用theta數列變化表達形狀）
theta = np.linspace(0, 4 * np.pi, 100)

# 設定初始半徑 a
a = 1

# 計算對應的 r
r = a * np.exp(b * theta)

# 轉換為 x, y 座標
x = r * np.cos(theta)
y = r * np.sin(theta)

# 繪製黃金螺旋
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='gold', linewidth=3)
plt.axis("equal")
plt.title("黃金螺旋")
plt.show()
```
