#Cross Entropy
因為我們只需要在意正解。所以如果訊息出現機率和真實世界的機率相似，我們就會覺得是對的。反之遇到不合常理的機率我們就扣分。


---


**<h3>資訊量(驚訝指數)</h3>**  

機率越高 越習以為常   
機率越低 越訝異   
所以把機率壓到0-1之間的話    
P接近1要很小(習以為常)   
P接近0要很大(訝異)   
log 能做到類似功能

```python
import numpy as np
import matplotlib.pyplot as plt

# 產生 x 值（避免 log(0) 的問題）
x = np.linspace(0.01, 1, 100)  # 從 0.01 到 1 之間取 100 個點
y = np.log(x)  # 計算 log(x)

# 畫圖
plt.figure(figsize=(6, 4))
plt.plot(x, y, label=r'$y = \log(x)$', color='b')

# 添加參考線
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)  # x 軸
plt.axvline(1, color='gray', linestyle='--', linewidth=0.8)  # y 軸交點

# 設定標籤與標題
plt.xlabel("x")
plt.ylabel("log(x)")
plt.title("Logarithm Function (0 - 1)")
plt.legend()
plt.grid(True)

# 顯示圖形
plt.show()
```

但因為log剛好相反，所以加個負號   
得到 $I(X) = -ln(P)$

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義 p 的範圍（從 0.1 到 1）
p = np.linspace(0.1, 1, 100)  # 生成 100 個點
negative_log_p = -np.log(p)  # 計算 -ln(p)

# 繪製圖形
plt.figure(figsize=(6, 4))
plt.plot(p, negative_log_p, label=r'$-\ln(p)$', color='b')

# 設定標籤與標題
plt.xlabel("p")
plt.ylabel(r"$-\ln(p)$")
plt.title("Plot of $-\ln(p)$")

# 添加網格與圖例
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# 顯示圖形
plt.show()
```

**<h3>結論</h3>**

所以我們在分類問題中ex正解為$[0,0,1]$   
把正確答案(yi)每一項乘上神經網路的那項y資訊量

得到以下結果
   
$
H(P, Q) = -\sum_{i=1}^{n} y_i \ln \hat{y}_i
$

#KL divergence


KL 散度我理解比較像是cross entropy跟真實世界的距離
**<h3>公式</h3>**

$D_{KL}(P||Q) = H(P,Q) - H(P)$   
$\uparrow$KL divergence  $\uparrow$最小值為$H(P)$$\uparrow$真實世界的entropy   
所以如果   
cross entropy$\uparrow$   
KL divergence$\uparrow$
